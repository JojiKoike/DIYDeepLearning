```python
import numpy as np
import matplotlib.pylab as plt
from typing import Callable, Tuple


def function_1(x: np.ndarray)-> np.ndarray:
    return 0.01 * x**2 + 0.1 * x + x**3


def contact_line(a: float, x_0: Tuple[float, float], x: np.ndarray) -> np.ndarray:
    return a*(x - x_0[0]) + x_0[1]


def numerical_diff(f: Callable[[np.ndarray], np.ndarray], x: np.ndarray) -> np.ndarray:
    dh: float = 1e-4
    return (f(x + dh) - f(x - dh)) / (2.0 * dh)


x: np.ndarray = np.arange(0, 20.0, 0.1)
y: np.ndarray = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y)

a = numerical_diff(function_1, np.array([5, 10]))

x_0: float = 5.0
y_0: float = function_1(np.array([x_0]))[0]
y_1: np.ndarray = contact_line(a[0], (x_0, y_0), x)
plt.plot(x, y_1)

x_0_1: float = 10.0
y_0_1: float = function_1(np.array([x_0_1]))[0]
y_2: np.ndarray = contact_line(a[1], (x_0_1, y_0_1), x)
plt.plot(x, y_2)

plt.show()
```

```python
from mpl_toolkits.mplot3d import  Axes3D
import matplotlib.pylab as plt
import numpy as np


def function_2(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    return x**2 + y**2


x: np.ndarray = np.arange(-5.0, 5.0, 0.1)
y: np.ndarray = x
X, Y = np.meshgrid(x, y)
Z = function_2(X, Y)

fig = plt.figure()
ax = Axes3D(fig)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

ax.plot_wireframe(X, Y, Z)
plt.show()
```
