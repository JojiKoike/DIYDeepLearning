```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 6, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--", label="cos")
plt.title("sin & cos")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
